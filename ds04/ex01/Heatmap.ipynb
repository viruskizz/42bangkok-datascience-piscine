```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
test_filename = os.path.join(os.getcwd(), 'ds04/assets/Test_knight.csv')
test_df = pd.read_csv(test_filename)
train_filename = os.path.join(os.getcwd(), 'ds04/assets/Train_knight.csv')
train_df = pd.read_csv(train_filename)
train_df.set_index('knight')
train_df.head(3)
```

```python
df = train_df.copy()
df['knight'] = df['knight'].map({'Jedi': 1, 'Sith': 0})
cor = df.corr()
cor
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(cor)
plt.title('Correlation Coefficient', fontsize=17, pad=20)
plt.gca().xaxis.set_label_position('top') 
plt.gca().xaxis.tick_bottom()
plt.show()
```
