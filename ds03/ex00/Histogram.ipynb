```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
test_filename = os.path.join(os.getcwd(), 'ds03/assets/Test_knight.csv')
test_df = pd.read_csv(test_filename)
train_filename = os.path.join(os.getcwd(), 'ds03/assets/Train_knight.csv')
train_df = pd.read_csv(train_filename)
train_df.set_index('knight')
```

```python
print('count:', len(test_df))
test_df.head(3)
```

```python
df = test_df
skills = df.columns
n_cols = 5
n_rows = len(skills) // n_cols
plt.figure(figsize=(20, 20))

for idx, skill in enumerate(skills):
    plt.subplot(n_rows, n_cols, idx + 1)
    df[skill].hist(bins=40, alpha=0.7, label='Knight', color='mediumseagreen')
    plt.title(skill)
    plt.legend()
plt.tight_layout()
plt.show()
```

```python
print('count:', len(train_df))
train_df.head(3)
```

```python
df = train_df
jedi_df = df[df['knight'] == 'Jedi']
sith_df = df[df['knight'] == 'Sith']
skills = df.columns[:-1] # Remove `knight` last index if existed
n_cols = 5
n_rows = len(skills) // n_cols
plt.figure(figsize=(20, 20))

for idx, skill in enumerate(skills):
    plt.subplot(n_rows, n_cols, idx + 1)
    jedi_df[skill].hist(bins=40, alpha=0.7, label='Jedi', color='royalblue')
    sith_df[skill].hist(bins=40, alpha=0.7, label='Sith', color='lightcoral')
    plt.title(skill)
    plt.legend()
plt.tight_layout()
plt.show()
```
